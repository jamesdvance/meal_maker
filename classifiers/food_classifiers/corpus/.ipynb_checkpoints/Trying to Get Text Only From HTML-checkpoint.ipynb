```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
pg = requests.get('https://www.allrecipes.com/')
pg_soup = BeautifulSoup(pg.text, 'html.parser')
```

```python
page_text = pg_soup.findAll(text=True)
```

```python
corpus=[]
for page in page_text:
    corpus.append(page.strip())
```

```python
def visible(element):
    if element.parent.name in ['style', 'script', '[document]', 'head', 'title']:
        return False
    elif re.match('<!--.*-->', str(element.encode('utf-8'))):
        return False
    else:
        return True
```

```python
corpus = filter(visible, page_text)
```

```python
txt = list(corpus)
```

```python
new_txt = []
for t in txt:
    new_txt.append(t.strip())
```

```python
new_txt
```

```python
corpus_par =''
for i in new_txt:
    corpus_par = corpus_par + ' '+i
```
