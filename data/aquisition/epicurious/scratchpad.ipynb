## Scrape A Page

```python
def scrape_epicurious_page(url):
    page = requests.get(url)

    food_pages_soup = BeautifulSoup(page.text, 'html.parser')
    description_dict = json.loads(food_pages_soup.find("meta",attrs={"name":"parsely-metadata"}).get("content"))
    # urllib.request.urlretrieve(description_dict["image-16-9"], description_dict["description"]+"-16-9.png")
    # urllib.request.urlretrieve(description_dict["image-1-1"], description_dict["description"]+"-1-1.png")
    steps = [child.text for child in food_pages_soup.find("div",attrs={"data-testid":"InstructionsWrapper"}).findChildren("p")]
    ingredients = [child.text for child in food_pages_soup.find("div",attrs={"data-testid":"IngredientList"}).findChildren("div")]
    name = food_pages_soup.find("h1",attrs={"data-testid":"ContentHeaderHed"}).text 
    ret_obj = {"name":name,
        "steps":steps,
        "ingredients":ingredients}
    ret_obj.update(description_dict)
    return ret_obj
```

## Crawl Site

```python
def crawl_subpages(main_page):
    output = [] 
    page = requests.get("https://www.epicurious.com/recipes-menus/most-saved-recipes")
    food_pages_soup = BeautifulSoup(page.text, 'html.parser')
    links = [page["href"] for page in food_pages_soup.findAll("a",attrs={"aria-label":"Get This Recipe"})]
    more_links = [page["href"] for page in food_pages_soup.findAll("a",attrs={"class":"view-complete-item"})]
    print(f"total links {len(links+more_links)}")
    for link in links+more_links:
        output.append(scrape_epicurious_page(link))

    return output
```
