Getting Relevant Metadata

```python
urls_df = pd.read_csv("food_related_urls.csv")
```

For each raw url, get metadata

For each domain / subdomain, get total number of pages

```python
base_url = 'http://index.commoncrawl.org/CC-MAIN-2018-09-index?url='
```

```python
urls_df['domain_pages'] = 0
for i in range(len(urls_df)-1):
#for i in range(10):
    domain = urls_df.iloc[i,1]
    full_req = base_url+domain+'/*&showNumPages=true&output=json&limit=1'
    response = requests.get(full_req)
    try:
        urls_df['domain_pages'][i] = response.json()['pages']
    except:
        pass
```

```python
urls_df.sort_values(by=['domain_pages'],axis=0, ascending=False)
```

```python
urls_df.columns.values
```

Creating a final url list

```python
corpus_urls = []
for i in range(len(urls_df)-1):
    corpus_urls.append(urls_df.urls[i])
    if urls_df.domain_pages[i] < 16 and urls_df.domain[i].find('.edu') == -1 and urls_df.domain[i].find('news') == -1 and urls_df.domain[i].find('post') == -1 and urls_df.domain[i].find('.vn') == -1:
# Should return language from the index as well if possible
        corpus_urls.append(urls_df.domain[i])
```

```python
corpus_df = pd.DataFrame(corpus_urls, columns=['urls'])
```

```python
corpus_df.to_csv("corpus_urls.csv", index=False)
```
